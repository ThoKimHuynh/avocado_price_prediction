```python
#Định nghĩa hàm tìm tổng số outliers của một thuộc tính
def findoutliers(df_frame, featurename):
    number_outliers = 0
    number_outliers = df_frame[df_frame[featurename] > (np.percentile(df_frame[featurename],75)
                                                      + 1.5*scipy.stats.iqr(df_frame[featurename]))].shape[0]
    number_outliers += df_frame[df_frame[featurename] < (np.percentile(df_frame[featurename],25)
                                                      - 1.5*scipy.stats.iqr(df_frame[featurename]))].shape[0]
    return number_outliers
```

```python
#Tìm các outliers của từng thuộc tính
print('Tìm số outliers cho từng thuộc tính của dữ liệu:\n')
number_outliers_tol = 0

#Sử dụng vòng lặp for để tính số outliers cho từng thuộc tính và số lượng outliers tổng cộng
for item in num_list:
    n = findoutliers(df, item)
    print('Số outliers của thuộc tính %s là: %d' %(item, n))
    number_outliers_tol += n
print('Tổng số outliers cho các thuộc tính là:', number_outliers_tol)
#print('Phần trăm outliers so với số lượng mẫu là: %.3f' % (number_outliers_tol/df.shape[0]))
```

### 3.2.2 Phân tích mối quan giữa các biến phân loại


```python
# Phân tích mối quan hệ giữa các cặp thuộc tính categorial
cat_list = ['Date', 'type', 'year', 'region', 'Month']
prob = 0.95
alpha = 1 - prob

for i in range(0, len(cat_list)-1):
    for j in range(i+1, len(cat_list)):
        
        two_way_table = pd.crosstab(df[cat_list[i]], df[cat_list[j]])        
        stat, p, dof, expected = chi2_contingency(two_way_table)        
        critical = chi2.ppf(prob, dof)
        
        if p <= alpha:
            print('Phân tích %s và %s:' %(cat_list[i], cat_list[j]))
            print('probability=%.3f, critical=%.3f, stat=%.3f, alpha=%.3f, p-value=%.3f' %(prob, critical, stat, alpha,p))
            print('%.3f <= %.3f -> Hai biến %s và %s phụ thuộc (Reject H0)' %(p, alpha, cat_list[i], cat_list[j]))
            
```

#### Nhận xét:
- Ta thấy các biến 'Date', 'Month' và 'year' phụ thuộc nhau, cho nên ta sẽ xem xét bỏ bớt biến 'Date' ra trong quá trình xây dựng model về sau.
- Thuộc tính 'type' và 'region' thì độc lập lẫn nhau.

Như phân tích trong phần trên, ta thấy AveragePrice có sự thay đổi theo thời gian. Vì vậy ta có thể phân tích thêm liệu AveragePrice có thay đổi theo các mùa trong năm hay không.

#### Tạo thêm cột 'season' 
- US's spring months are March, April and May(3,4,5)
- US's summer months are June, July, and August (6,7,8)
- US's autumn months are September, October, and November (9,10,11)
- US's winter months are December, January, and February (12,1,2)

```python
def convert_month(month):
    if month == 3 or month == 4 or month == 5:
        return 0
    elif month == 6 or month ==7 or month == 8:
        return 1
    elif month == 9 or month == 10 or month ==11:
        return 2
    else:
        return 3
```

```python
#Tạo thêm côt 'season'
df['season'] = df['Month'].apply(lambda x: convert_month(x))
```

## Bước 4: Xây dựng Models dự báo cho bài toán

- Các thuộc tính được chọn để đưa vào xây dựng mô hình để dự đoán AveragePrice là: Total Volume, year, Month, type, region.

** Xây dựng Pipeline để tái sử dụng cho 1 thuật toán về sau khi cần

```python
cat_names = X.select_dtypes('object').columns.tolist()
num_names = X.select_dtypes(['float','int']).columns.tolist()
```

```python
#Xây dựng pipeline cho thuộc tính phân loại
cat_pl = Pipeline(
    steps=[('onehot', OneHotEncoder(drop='first'))]
)
```

```python
#Xây dựng pipeline cho thuộc tính liên tục - Sử dụng RobustScaler vì có outliers
num_pl = Pipeline(
    steps=[('scaler', RobustScaler())]
)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pl, num_names), #áp dụng cho thuộc tính liên tục
        ('cat', cat_pl, cat_names), #áp dụng cho thuộc tính phân loại
    ]
)
```

```python
#Xây dựng hàm tìm giá trị K tối ưu cho thuật toán KNN Regression
def optimal_k(X_train, X_test, y_train, y_test):
    list_k =[]
    list_r_score =[]

    for K_value in range(2,11): 
      list_k.append(K_value)
      neigh = KNeighborsRegressor(n_neighbors=K_value)
      neigh.fit(X_train,y_train)
      score = neigh.score(X_test,y_test)*100
      list_r_score.append(score)
      print('The R^2 score is: ', score, ' for K-Value:',K_value)

    vi_tri = list_r_score.index(max(list_r_score))
    k = list_k[vi_tri]

    print('\nThe optimal number of neighbors is:', k, 'with R^2 score is:', list_r_score[vi_tri])
    return k
```

```python
# Xây dựng hàm tính độ chính xác train và test acc theo các mô hình sau: 
# LinearRegression, SVR, RandomForestRegressor, DecisionTreeRegressor
# KNeighborsRegressor, XGBRegressor, 
def compute_models(X_train, X_test, y_train, y_test, k):
    models = [
              LinearRegression(),
              SVR(),
              RandomForestRegressor(n_estimators=100),
              XGBRegressor(random_state=0),
              DecisionTreeRegressor(),
              KNeighborsRegressor(n_neighbors=k),



    ]
    CV = 5 # Số lần chạy cho từng thuật toán
    entries =[]
    #i=0
    for model in models:
      scores_train = []
      scores_test = []
      abs_scores = []
      times = []
      for i in range(CV):
        t1 = datetime.datetime.now()
        model_name = model.__class__.__name__
        model.fit(X_train,y_train)
        t2 = datetime.datetime.now()
        d = round((t2-t1).microseconds/1000,1) #chuyển sang miliseconds
        
        score_train = model.score(X_train,y_train)
        score_test = model.score(X_test,y_test)
        abs_score = abs(score_train - score_test)

        scores_train.append(score_train)
        scores_test.append(score_test)
        abs_scores.append(abs_score)
        times.append(d)


      print(model.__class__.__name__, scores_test)
      entries.append([model_name,np.array(scores_train).mean(),
                      np.array(scores_test).mean(),np.array(abs_scores).mean(),
                      np.array(times).mean()])

      #i = i+1
    cv_df = pd.DataFrame(entries,
                         columns=['model_name','score_train_mean',
                                   'score_test_mean','abs|score|','time_mean'])

    return cv_df
```

### 4.1 - Trường hợp 1: Ta xây dựng mô hình dựa trên dữ liệu gốc, không xử lý outliers

```python
#Thực hiện encoding và scaling cho dữ liệu đầu vào - sử dụng pipeline 'preprocessor'
encode_scaling = preprocessor.fit(X)
processed_X = encode_scaling.transform(X)
```

```python
#Chia bộ dữ liệu thành train, test theo tỉ lệ 70:30
X_train, X_test, y_train, y_test = train_test_split(processed_X,y,
                                                       test_size=0.3,
                                                       random_state=0)
```

```python
#Tìm k tối ưu của KNN
k = optimal_k(X_train, X_test, y_train, y_test)
```

```python
#Tính toán và so sánh các thuật toán
cv_df = compute_models(X_train, X_test, y_train, y_test, k)
```

#### Nhận xét:

- DecisionTree thì bị overfitting.
- KNN thì cho R2 score của test và train cao và chênh lệch ít, nhưng lại thấp hơn Random Forest.
- Random Forest thì cho R2 cao nhất, nhưng có khuynh hướng hơi bị overfitting.
